```python
import random
import math
```

```python
def avg(lis):
  sum=0
  for i in range(len(lis)):
    sum=sum+lis[i]
  return sum/len(lis)
avg([2,3,4])
```

This code will find average distance between 2 points in 1* 1* 1*....(N dimension) hypercube(Hypercube with all adges equal to 1) 

```python
import math
def rand(iteration,dim):
  lstofrandom=[]
  d=0
  count=0
  for i in range(0,iteration):
    d=0
    for i in range(0,dim):
        n1=random.random()
        n2=random.random()
        d=d+(n1-n2)**2
    lstofrandom.append(math.sqrt(d))
  return avg(lstofrandom)
```
