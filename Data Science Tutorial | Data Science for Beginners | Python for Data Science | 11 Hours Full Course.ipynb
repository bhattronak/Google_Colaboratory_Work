COVARIANCE.PNG

1. Formula in above pic shows covariance.
If its +ve that does mean both X & Y follow same path or moving in same direction and if its -ve then X & Y are moving in opposite directions.

2. Value of corelation is harder to interpretate because of diffrent units of X & Y. So to avoid this problem we divide both X & Y with their S.D to make formula unit free.

3. After dividing above formula with S.D we get new term called correlation, which will tell relation between X & Y.

4. Range of correlation is -1<=correlation<=1.

5. If correlation is 1 that does mean both X & Y are very relatable to each other.

6. If correlation is -1 that does mean both X & Y are not following each other at all.

7. If correlation is 0 then there is no relation between 2 variables.

8. Correlation is also known as R value

9. R value>0.7 or R value<-0.7 is consider as a good R value 

10. If 0.5>R value>0.7 or -0.7< R value<-0.5,It requires further research on data.

11. Any R value close to 0, Not worth to consider.
